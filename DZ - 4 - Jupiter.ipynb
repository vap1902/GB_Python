```python
c_m = df.corr()
c_m = np.round(c_m,1)
c_m[np.abs(c_m) < 0.3] = 0
```

```python
plt.figure(figsize=(10,8))
sns.set(font_scale=1.4)
sns.heatmap(c_m, annot=True, linewidths=0.5, cmap = 'coolwarm')
plt.title('Корреляционная матрица')
```
