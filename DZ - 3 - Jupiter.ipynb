## 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
df['delta_renovated1'] = df['yr_renovated'] - df['yr_built']
```

```python
def zero_repl(x):
    if x < 0:
        x = 0
    return x
df['delta_renovated'] = df['delta_renovated1'].apply(zero_repl)
df.drop(columns='delta_renovated1', inplace = True)
df
```
